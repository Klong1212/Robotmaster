```python
import pandas as pd

# อ่านข้อมูลจากไฟล์ CSV ที่เก็บข้อมูลเซ็นเซอร์ TOF และ Gimbal
df = pd.read_csv('square_pid_log_with_xy_15_005_03.csv')

# กรองข้อมูลที่มีระยะทาง TOF มากกว่า 0 และน้อยกว่า 500 mm เพื่อกำจัดค่าผิดปกติ
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = df['current_x']
y = df['current_y']

# แสดงผลเป็นจุด
plt.figure(figsize=(6, 6))
plt.scatter(y, x, s=10, color='blue',marker='x')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid()
plt.gca().set_aspect('equal') # เพื่อให้แกน X และ Y มีสเกลเท่ากัน ทำให้กราฟดูไม่บิดเบี้ยว
plt.title("2D Scan Points from position_x and position_y")
plt.xlabel("X (mm)")
plt.ylabel("Y (mm)")
plt.show()
```
