```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import butter, filtfilt

# โหลดข้อมูล
df3 = pd.read_csv("toffile.csv")
```

```python
# ดึงเฉพาะ tof_distance_mm (ระยะ ToF)
tof = df3["tof_distance_mm"].dropna().astype(float).reset_index(drop=True)

# Moving Average filter (window=5)
window = 3
ma_filter = tof.rolling(window=window, center=True).mean()

# Plot raw ToF และ Moving Average
plt.figure(figsize=(16,12))
plt.subplot(2, 2, 1)
plt.plot(tof, label="Raw ToF (mm)", alpha=0.6)
plt.xlabel("Samples")
plt.ylabel("Distance (mm)")
plt.title("Raw ToF Distance ")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(16,12))
plt.subplot(2, 2, 2)
plt.plot(ma_filter, label=f"Moving Average (window={window})", linewidth=2, color='y')
plt.xlabel("Samples")
plt.ylabel("Distance (mm)")
plt.title("ToF Distance Moving Average Filter")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(16,12))
plt.subplot(2, 2, 3)
plt.plot(tof, label="Raw ToF (mm)", alpha=0.6)
plt.plot(ma_filter, label=f"Moving Average (window={window})", linewidth=2, color='y')
plt.xlabel("Samples")
plt.ylabel("Distance (mm)")
plt.title("ToF Distance vs Moving Average Filter")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ดึงเฉพาะ tof_distance_mm (ระยะ ToF)
tof = df3["tof_distance_mm"].dropna().astype(float).reset_index(drop=True)

# Exponential Moving Average filter (span = window)
window = 3
ema_filter = tof.ewm(span=window, adjust=False).mean()

# Plot raw ToF
plt.figure(figsize=(16,12))
plt.subplot(2, 2, 1)
plt.plot(tof, label="Raw ToF (mm)", alpha=0.6)
plt.xlabel("Samples")
plt.ylabel("Distance (mm)")
plt.title("Raw ToF Distance")
plt.legend()
plt.grid(True)
plt.show()

# Plot EMA
plt.figure(figsize=(16,12))
plt.subplot(2, 2, 2)
plt.plot(ema_filter, label=f"Exponential Moving Average (span={window})", linewidth=2, color='orange')
plt.xlabel("Samples")
plt.ylabel("Distance (mm)")
plt.title("ToF Distance Exponential Moving Average Filter")
plt.legend()
plt.grid(True)
plt.show()

# Plot Raw vs EMA
plt.figure(figsize=(16,12))
plt.subplot(2, 2, 3)
plt.plot(tof, label="Raw ToF (mm)", alpha=0.6)
plt.plot(ema_filter, label=f"EMA (span={window})", linewidth=2, color='orange')
plt.xlabel("Samples")
plt.ylabel("Distance (mm)")
plt.title("ToF Distance vs EMA Filter")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ดึงเฉพาะ tof_distance_mm (ระยะ ToF)
tof = df3["tof_distance_mm"].dropna().astype(float).reset_index(drop=True)

# Median Filter (ME)
window = 3
me_filter = tof.rolling(window=window, center=True).median()

# Plot raw ToF
plt.figure(figsize=(16,12))
plt.subplot(2, 2, 1)
plt.plot(tof, label="Raw ToF (mm)", alpha=0.6)
plt.xlabel("Samples")
plt.ylabel("Distance (mm)")
plt.title("Raw ToF Distance")
plt.legend()
plt.grid(True)
plt.show()

# Plot ME
plt.figure(figsize=(16,12))
plt.subplot(2, 2, 2)
plt.plot(me_filter, label=f"Median Filter (window={window})", linewidth=2, color='green')
plt.xlabel("Samples")
plt.ylabel("Distance (mm)")
plt.title("ToF Distance Median Filter")
plt.legend()
plt.grid(True)
plt.show()

# Plot Raw vs ME
plt.figure(figsize=(16,12))
plt.subplot(2, 2, 3)
plt.plot(tof, label="Raw ToF (mm)", alpha=0.6)
plt.plot(me_filter, label=f"Median Filter (window={window})", linewidth=2, color='green')
plt.xlabel("Samples")
plt.ylabel("Distance (mm)")
plt.title("ToF Distance vs Median Filter")
plt.legend()
plt.grid(True)
plt.show()
```
